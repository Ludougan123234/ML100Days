# Introduction to RNN

- What is RNN? 
  - RNN is a class of neural network that takes the output of the previous layer in a neural network as inputs for the next layer of the neural network 
  - A typical RNN model architechture is shown below: 
  ![RNN](https://stanford.edu/~shervine/teaching/cs-230/illustrations/architecture-rnn-ltr.png?9ea4417fc145b9346a3e288801dbdfdc)
  where the timestamp $t$, activation $a^t$, and output $y^t$ are expressed as: 
  $\begin{gather} a^{<t>} = g_1(W_{aa}a^{<t-1>} + W_{ax}x^{<t>} + b_a) \\ \text{and} \\ y^{<t>} = g_2(W_{ya}^{<t>} + b_y) \end{gather} \\ \text{with } W_{ax}, W_{aa}, W_{ya}, b_a, b_y \text{ coefficients that are shared teporally, and } g_1, g_2 \text{ are activation functions}$ ![rnn2](https://stanford.edu/~shervine/teaching/cs-230/illustrations/description-block-rnn-ltr.png?74e25518f882f8758439bcb3637715e5)
  


- Loss function of RNN: $\mathcal{L}(\hat{y}, y) = \sum^{T_y}_{t=1}\mathcal{L}(\hat{y}^{<t>}, y^{<t>})$
- Backpropagation through timestamps: $\frac{\partial\mathcal{L}}{\partial W} = \sum^{T}_{t=1}\frac{\partial\mathcal{L}^{(t)}}{\partial W} _{(t)}$

# Problems with RNN: 

- Problem: Exploding/vanishing gradient: 
  - A gradient is a vector that is used to optimize a neural network by updating the network weights in the appropriate direction with the right amount. 
  - When training an RNN, the gradient can "explode", meaning that the gradient gets so large that it makes a network unstable (poor or large changes in loss), or even overflow available memory and result in NaN values.
  - The gradient can also "vanish", meaning that the values in the gradient get infinitely close to zero, leaving the initial or lower layers unchanged and resulting in a process of gradient descent that fails to converge to the optimum. The tell-tale signs of a vanishing gradient include - 1) 0 model weight, 2) learning becomes very slow, and 3) parameters in higher layers change drastically while no changes can be observed in lower layers. 

- Solutions: 
  1. Gradient clipping: Achieved by capping the maximum value for the gradient
  2. Adopt non-saturating activation functions: `sigmoid` and `tanh` are saturating activation functions, meaning that the function "squeezes" the input into a specific range (sigmoid: $[0,1]$, tanh: $[-1,1]$). On the contrary, a function is non-saturating iff $\lim_{z \rightarrow \infty} f(z) = \infty$. Typical examples of non-saturating activation functions include: `ReLU`, `Leaky ReLU`, and `ELU`
  3. Batch normalization: Re-center and re-scale the inputs
  4. Use GRU (Gated Recurrent Unit) or LSTM (Long Short-Term Memory): by using gates, vanishing gradient can be solved.
