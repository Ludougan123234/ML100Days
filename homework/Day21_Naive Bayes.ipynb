## Problem: Why does the assumption of conditional independence make naïve bayes "naïve"?

###Answer: "conditional independence" 

According to Scikit learn documentation, Naive Bayes is an algorithm that applies Bayes' theorem with the "naive" assumption "of conditional independence between every pair of features."

The reason why Naïve Bayes is called "naïve" is that the assumption of conditional independence is too strong and unrealistic (i.e., almost never true) for most datasets (Shobha and Rangaswamy, 2019). However, Naïve Bayes is still very useful for large datasets.
