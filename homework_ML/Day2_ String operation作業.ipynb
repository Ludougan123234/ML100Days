### isnumeric(), isdigit(), isdecimal() 各有幾個
<font color='green'>Count the number of `isnumeric()`, `isdigit()`, and `isdecimal()`</font>

```python
test_string = ['5.9', '30', '½', '³', '⑬']
```

```python
def spam(s, isnumeric_count, isdigit_count, isdecimal_count):
    for attr in ['isnumeric', 'isdecimal', 'isdigit']:
        if attr == 'isnumeric':
            if s.isnumeric():
                isnumeric_count += 1
        if attr == 'isdecimal':
            if s.isdecimal():
                isdecimal_count += 1
        if attr == 'isdigit':
            if s.isdigit():
                isdigit_count += 1
    return isnumeric_count, isdecimal_count, isdigit_count
                      
(isnumeric_count, isdigit_count, isdecimal_count) = (0,0,0)
for i in test_string: 
    isnumeric_count, isdigit_count, isdecimal_count = spam(i, isnumeric_count, isdigit_count, isdecimal_count)
```

```python
print('isnumeric_count: {}'.format(isnumeric_count))     
print('isdigit_count: {}'.format(isdigit_count))     
print('isdecimal_count: {}'.format(isdecimal_count))  
```

## 依照指示轉換number output format

<font color='green'>Convert number to designated format</font>

```python
number = 3.1415926
```

#### 轉換為科學符號表示法 (小數點後兩位)

<font color='green'>Instruction 1: Round the decimal number to 2 decimal places</font>

```python
print("{}".format("%.2e"%number))
```

#### 轉換為%
<font color='green'>Convert the number to percentage format</font>

```python
print('{:.2%}'.format(number))
```

#### 補零成為3.14159300

<font color='green'>Zero-padding</font>

```python
print("{:0<10}".format("%f"%number))
```
